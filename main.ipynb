```python
# Use the data from https://www.nycbynatives.com/nyc_info/new_york_city_zip_codes.php to get zipcodes for each borough
zipdata = pd.read_csv('data/zip.csv')
zipdf = zipdata.groupby('borough')['zip'].apply(list).reset_index(name='zipcodes')
zipdf['borough'] = zipdf['borough'].str.upper()
zipdf.loc[zipdf['borough'] == 'STATEN', 'borough'] = 'STATEN ISLAND'

# Create Valid zipcodes in NYC by combining all zipcodes of the 5 boroughs
valid_zipcodes = []
for lst in zipdf['zipcodes']:
    valid_zipcodes += lst
```

```python
# Fill in 'Unsepcified' for the N/A's in the Borough column
df_2020['Borough'] = df['Borough'].fillna('Unspecified')

# Remove N/A zipcodes and invalid zipcodes
df_2020 = df_2020[df_2020['Incident Zip'].notna()]
df_2020 = df_2020[df_2020['Incident Zip'].isin(valid_zipcodes)]
```

```python
# Clean up 'Unspecified' boroughs by looking up the zipcodes and fit them to their corresponding borough
for idx in range(len(zipdf)):
    df_2020.loc[(df_2020['Borough'] == 'Unspecified') &
                (df_2020['Incident Zip'].isin(zipdf['zipcodes'][idx])), 'Borough'] = zipdf['borough'][idx]
```

```python
# Change type of Incident Zip to int
df_2020['Incident Zip'] = df_2020['Incident Zip'].astype('int')
```

```python
# Make sure we have cleaned up the Unspecified ones
print('List of Boroughs after cleaning')
set(df_2020['Borough'])
```

# Goal1: Consider only the 10 most common overall complaint types. For each borough, how many of each of those 10 types were there in 2020?

```python
# Select cases that includes the top 10 complaint types only
top10_complaint_df = df_2020.loc[df_2020['Complaint Type'].isin(Top10_Type)]

# List of boroughs
boroughs = list(set(df_2020['Borough']))

print("Boroughs: " + str(boroughs))
for borough in boroughs:
    temp_df = top10_complaint_df[top10_complaint_df['Borough'] == borough]
    lst = temp_df.groupby('Complaint Type').size().sort_values(ascending=True)
    index_lst = lst.index.values.tolist()
    plt.figure(figsize=(10,5))
    plt.barh(index_lst,list(lst))
    for i, v in enumerate(list(lst)):
        plt.text(v, i,str(v))
    plt.title(borough + "'s Number of Complaint Types")
    plt.show()
```

### Notice: Noise - Residential is the most complaint type accross boroughs and Street Light Condition is the least

# Goal 2. Consider only the 10 most common overall complaint types. For the 10 most populous zip codes, how many of each of those 10 types were there in 2020?
Assuming ZCTA = zip and 2010 census population by zipcode data represents the data in 2020

```python
# Read population zipcode data
zip_df = pd.read_csv('data/zipcode_population.csv')

# Filter 10 most populous zipcode
zip_df = zip_df[(zip_df['Zip Code ZCTA'] > 10000) & (zip_df['Zip Code ZCTA'] <= 11697)]
zip_df = zip_df.sort_values('2010 Census Population',ascending=False)
Top10_zipcode = list(zip_df['Zip Code ZCTA'].head(10))
print("Top 10 most populous zipcodes and populations")
zip_df.head(10)
```

```python
print("Top 10 zipcodes: " + str(Top10_zipcode))
for zipcode in Top10_zipcode:
    temp_df = top10_complaint_df[top10_complaint_df['Incident Zip'] == float(zipcode)]
    lst = temp_df.groupby('Complaint Type').size().sort_values(ascending=True)
    index_lst = lst.index.values.tolist()
    plt.figure(figsize=(10,5))
    plt.barh(index_lst,list(lst))
    for i, v in enumerate(list(lst)):
        plt.text(v, i,str(v))
    plt.title("Zipcode " + str(zipcode) + "'s Number of Complaint Types")
    plt.show()
```

### Notice: The most complaint type accross the zipcodes is Noise-Residential and the least one is Street Light Condition. However, zipcode 11368's most complaint type is Blocked Driveway and there's barely Street Light Condition in zipcode 10025 

# Goal3: Considering all complaint types. Which boroughs are the biggest "complainers" relative to the size of the population in 2020? Meaning, calculate a complaint-index that adjusts for population of the borough.

```python
'''
Getting population of each borough from https://en.wikipedia.org/wiki/Boroughs_of_New_York_City
Define complain index = number of complaints in that borough / population of that borough

Borough      |      Population(2020)
Bronx                  1,472,654
Brookyln               2,736,074
Manhattan              1,694,263
Queens                 2,405,464
Staten Islands          495,747
'''
population = [1472654, 2736074, 1694263, 2405464, 495747]
borough_list = ['BRONX', 'BROOKLYN', 'MANHATTAN', 'QUEENS', 'STATENISLAND']

pop_df = df_2020.groupby('Borough').size()
complaint_indexes = []
for idx in range(5):
    size = population[idx]
    cases = pop_df[idx]
    index = round(cases / size, 3)
    complaint_indexes.append(index)
complaint_indexes = sorted(complaint_indexes, reverse=True)
    
borough_list2 = ['BRONX', 'MANHATTAN', 'BROOKLYN', 'QUEENS', 'STATENISLAND']
    
plt.figure(figsize=(15,5))
plt.barh(borough_list2,complaint_indexes)
plt.gca().invert_yaxis()
for i, v in enumerate(complaint_indexes):
    plt.text(v, i,str(v))
plt.title("Complaint Index rankings")
plt.show()
```

### Result: The biggest complainer in NYC is BRONX by measuring the complaint indexes. The rest of them are Manhattan, Brooklyn, Queens, and Staten Island in decreasing order.
